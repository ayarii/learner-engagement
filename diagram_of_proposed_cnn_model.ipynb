```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Function to draw the arrow between blocks
def draw_arrow(x1, y1, x2, y2):
    plt.annotate('', xy=(x2, y2), xytext=(x1, y1),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8))

# Function to draw a box with rounded corners
def draw_box(x, y, width, height, text, color='lightgray'):
    box = FancyBboxPatch((x, y), width, height, boxstyle="round,pad=0.1", edgecolor='black', facecolor=color)
    plt.gca().add_patch(box)
    plt.text(x + width/2, y + height/2, text, ha="center", va="center", fontsize=12, weight='bold')

# Initialize plot
plt.figure(figsize=(10, 8))
plt.axis('off')

# Draw custom CNN blocks
draw_box(0.1, 0.7, 0.3, 0.1, "Conv Layer (32) -> MaxPool", color="lightblue")
draw_box(0.1, 0.55, 0.3, 0.1, "Conv Layer (64) -> MaxPool", color="lightblue")
draw_box(0.1, 0.4, 0.3, 0.1, "Conv Layer (128) -> MaxPool", color="lightblue")

# Draw Pre-trained VGG16 block
draw_box(0.6, 0.55, 0.3, 0.2, "Pre-trained VGG16", color="lightgreen")

# Draw Concatenation Layer
draw_box(0.35, 0.25, 0.3, 0.1, "Concatenation Layer", color="lightyellow")

# Draw Dense Layer
draw_box(0.35, 0.1, 0.3, 0.1, "Dense Layer (512 units)", color="orange")

# Draw Output Layer
draw_box(0.35, -0.05, 0.3, 0.1, "Output Layer (SoftMax)", color="tomato")

# Draw arrows for Custom CNN Branch
draw_arrow(0.25, 0.75, 0.25, 0.65)  # Conv 32 -> Conv 64
draw_arrow(0.25, 0.6, 0.25, 0.5)  # Conv 64 -> Conv 128
draw_arrow(0.25, 0.45, 0.45, 0.45)  # Conv 128 -> Concatenation

# Draw arrows for Pre-trained VGG16 Branch
draw_arrow(0.75, 0.65, 0.75, 0.45)  # VGG16 -> Concatenation
draw_arrow(0.45, 0.35, 0.45, 0.25)  # Concatenation -> Dense
draw_arrow(0.45, 0.15, 0.45, 0.05)  # Dense -> Output

# Show the diagram
plt.show()
```
